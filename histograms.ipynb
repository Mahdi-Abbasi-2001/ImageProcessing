```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
horse = cv2.imread('img/horse.jpg')
show_horse = cv2.cvtColor(horse, cv2.COLOR_BGR2RGB)

bricks = cv2.imread('img/bricks.jpg')
show_bricks = cv2.cvtColor(bricks, cv2.COLOR_BGR2RGB)
```

```python
horse_hist_values = cv2.calcHist([horse], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
```

```python
horse_hist_values.shape
```

```python
plt.plot(horse_hist_values)
```

```python
bricks_hist_values = cv2.calcHist([bricks], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
```

```python
bricks_hist_values.shape
```

```python
plt.plot(bricks_hist_values)
```

```python
colors = ('b', 'g', 'r')

for i, color in enumerate(colors):
    histr = cv2.calcHist([bricks], [i], None, [256], [0, 256])
    plt.plot(histr, color=color)
    plt.xlim([0, 256])

plt.title("Histogram for rainbow")
```
