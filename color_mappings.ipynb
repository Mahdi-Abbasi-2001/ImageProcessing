```python
import cv2
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('img/perceptual-standard.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img)
```

```python
hls_img = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
plt.imshow(hls_img)
```
