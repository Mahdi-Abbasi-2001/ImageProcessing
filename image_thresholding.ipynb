```python
crossword_img = cv2.imread("img/crossword.jpg", 0)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap="gray")
```

```python
show_pic(crossword_img)
```

```python
crossword_ret_binary, crossword_thresh_binary = cv2.threshold(crossword_img, 155, 255, cv2.THRESH_BINARY)
show_pic(crossword_thresh_binary)
```

```python
crossword_adapt_thresh_binary = cv2.adaptiveThreshold(crossword_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(crossword_adapt_thresh_binary)
```

```python
crossword_img_blended = cv2.addWeighted(src1=crossword_thresh_binary, alpha=0.6, src2=crossword_adapt_thresh_binary, beta=0.4, gamma=0)
show_pic(crossword_img_blended)
```
