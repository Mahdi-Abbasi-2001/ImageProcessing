```python
import cv2
import matplotlib.pyplot as plt
```

```python
balloons_img = cv2.imread("img/perceptual-standard.jpg")
balloons_img = cv2.cvtColor(balloons_img, cv2.COLOR_BGR2RGB)
smile_img = cv2.imread("img/smile.jpg")
smile_img = cv2.cvtColor(smile_img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(balloons_img)
```

```python
plt.imshow(smile_img)
```

```python
balloons_img.shape
```

```python
smile_img.shape
```

First Option

```python
resized_smile_img = cv2.resize(smile_img, (2048, 1210))
```

```python
plt.imshow(resized_smile_img)
```

```python
blended_img = cv2.addWeighted(src1=balloons_img, alpha=0.8, src2=resized_smile_img, beta=0.2, gamma=0)
```

```python
plt.imshow(blended_img)
```

Second Option

```python
x_offset, y_offset = 0, 0
x_end, y_end = x_offset + smile_img.shape[1], y_offset + smile_img.shape[0]
```

```python
balloons_img_copy = balloons_img.copy()
```

```python
balloons_img_copy[y_offset:y_end, x_offset:x_end] = smile_img
```

```python
plt.imshow(balloons_img_copy)
```
