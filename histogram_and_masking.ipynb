```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainbow = cv2.imread('img/rainbow.jpg')
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
```

```python
rainbow.shape
```

```python
mask = np.zeros(shape=(rainbow.shape[:2]), dtype=np.uint8)
```

```python
mask[300:400, 100:400] = 255
plt.imshow(mask, cmap="gray")
```

```python
rainbow_masked = cv2.bitwise_and(rainbow, rainbow, mask=mask)
```

```python
show_rainbow_masked = cv2.bitwise_and(show_rainbow, show_rainbow, mask=mask)
plt.imshow(show_rainbow_masked)
```

```python
hist_mask_values_red = cv2.calcHist([rainbow], channels=[2], mask=mask, histSize=[256], ranges=[0, 256])
```

```python
hist_values_red = cv2.calcHist([rainbow], channels=[2], mask=None, histSize=[256], ranges=[0, 256])
```

```python
plt.plot(hist_mask_values_red)
plt.title("RED HISTOGRAM FOR MASKED RAINBOW")
```

```python
plt.plot(hist_values_red)
plt.title("RED HISTOGRAM FOR NORMAL RAINBOW")
```
